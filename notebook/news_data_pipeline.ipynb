```python
def scrape_content(driver, url):
    driver.get(url)
    time.sleep(2)

    try:
        headline = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//section[contains(@class, 'article') or contains(@class, 'narticle')]//h1"))
        ).text
    except:
        headline = "N/A"

    try:
        summary = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "h2[class='jsx-a1664bcd8826003c narticle-intext schema-summary-target']"))
        ).text
    except:
        summary = "N/A"

    try:
        datetime_element = driver.find_element(By.CSS_SELECTOR, "span[class='jsx-61453e8285f2673c pub-date']")
        published_time = datetime_element.text
    except:
        published_time = "N/A"

    try:
        timer = driver.find_element(By.CSS_SELECTOR, "span[class='jsx-1b99d3191c21f586 nart-time']")
        times = timer.text
    except:
        times = "N/A"

    return {
        "url": url,
        "headline": headline,
        "summary": summary,
        "published_time": published_time,
        "times": times
    }
```

```python
import math
def estimate_reading_time(text):
    words = text.split()
    return math.ceil(len(words) / 200)

def scrape_fe_b(driver, url):
    driver.get(url)

    try:
        headline = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "h1.heading-three"))).text

        try:
            summary = driver.find_element(By.CSS_SELECTOR, "h2.heading-four").text
        except:
            summary = ""

        try:
            date_time = driver.find_element(By.CSS_SELECTOR, "div.updated.flex time").text
        except:
            date_time = ""

        paragraphs = driver.find_elements(By.CSS_SELECTOR, "#pcl-full-content p")
        body = "\n".join([p.text for p in paragraphs if p.text.strip()])

        reading_time_min = estimate_reading_time(body)

        return {
            "url": url,
            "headline": headline,
            "summary": summary,
            "date_time": date_time,
            "body": body,
            "reading_time_min": reading_time_min
        }

    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return None
```

# Used VADER sentiment analysis to compute sentiment scores for Business news articles.  
# Classified articles as Positive, Negative, or Neutral based on compound scores.


```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Get sentiment scores
business['sentiment_score'] = business['combined'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

# Classify as Positive, Negative, Neutral
def get_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

business['sentiment_label'] = business['sentiment_score'].apply(get_sentiment)

business
```

# Applied TF-IDF to convert text data into numerical features and used NMF (Non-negative Matrix Factorization) to identify topics in Business and Market news datasets.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
#TF-IDF = Term Frequency - Inverse Document Frequency. It gives higher scores to words that are important but not too common.
# Business data TF-IDF
vectorizer_business = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
business_tfidf = vectorizer_business.fit_transform(business['combined'])

# Market data TF-IDF
vectorizer_market = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
market_tfidf = vectorizer_market.fit_transform(market['combined'])

# NMF for Business
n_topics = 6  # or 6, 7, etc.
nmf_business = NMF(n_components=n_topics, random_state=42)
business_topics = nmf_business.fit_transform(business_tfidf)

# NMF for Market
nmf_market = NMF(n_components=n_topics, random_state=42)
market_topics = nmf_market.fit_transform(market_tfidf)
```
